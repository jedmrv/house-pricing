```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import mean_squared_error
```

```python
Xtrain, Xtest, Ytrain, Ytest = tts(X, Y, test_size=0.2, random_state=0)
#xtrain, xtest, ytrain, ytest = tts(x, y, test_size=0.2, random_state=0)
```

```python
linear = LinearRegression(fit_intercept = True)
ridge = Ridge(fit_intercept = True)
lasso = Lasso(fit_intercept = True)
d = DecisionTreeRegressor()
g = GradientBoostingRegressor()
r = RandomForestRegressor()
k = KNeighborsRegressor(n_neighbors=5)
```

```python
reg = [linear,ridge,lasso,d,g,r,k]
regs = ['Linear','Ridge','Lasso','Decision Tree','Gradient Boosting','Random Forest','Kneighbors']
for i in range(len(reg)):
    reg[i] = reg[i].fit(Xtrain,Ytrain)
    print(regs[i])
    print('Rsquared training', reg[i].score(Xtrain,Ytrain))
    print('Rsquared testing', reg[i].score(Xtest,Ytest))
    print('Difference of Rsquared', (reg[i].score(Xtrain,Ytrain) - reg[i].score(Xtest,Ytest)))
    #print('Predict', reg[i].predict(Z))
    print('RMSE of training', mean_squared_error(Ytrain, reg[i].predict(Xtrain))**.5)
    print('RMSE of testing', mean_squared_error(Ytest, reg[i].predict(Xtest))**.5)
    #print('Diff of RMSE', (mean_squared_error(Ytrain, reg[i].predict(Xtrain))**.5)-(mean_squared_error(Ytest, reg[i].predict(Xtrain))**.5))
    
## We are trying to predict Ytrain and Xtrain, Ytest and Xtest for rmse
# linear, ridge and lasso - only check the accuracy of train
# score in classification algs is always accuracy
# rsquared for regression
# the reason why kneighbors is low is because its not imposing linear relationship, its trying to model grouping
```
